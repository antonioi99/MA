```python
def renumber_json_ids(input_path, output_path, start_id=12500):
    """
    Renumbers the top-level keys in a JSON file starting from `start_id`.

    Args:
        input_path (str): Path to the input JSON file.
        output_path (str): Path where the updated JSON should be saved.
        start_id (int): New starting index (default = 64).
    """
    # Load the JSON file
    with open(input_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Convert keys to integers and sort to ensure correct order
    old_keys = sorted(map(int, data.keys()))

    # Create new dictionary with updated keys
    new_data = {str(i + start_id): data[str(old_key)] for i, old_key in enumerate(old_keys)}

    # Save updated JSON
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(new_data, f, ensure_ascii=False, indent=4)

    print(f"✅ Renumbered keys from {old_keys[0]}–{old_keys[-1]} to {start_id}–{start_id + len(old_keys) - 1}")
    print(f"💾 Saved updated file to: {output_path}")
```
